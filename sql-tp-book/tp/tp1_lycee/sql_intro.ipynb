Il existe deux manières de taper des lignes de commandes SQL dans un Jupyter Notebook : 
- avec des décorateurs pour taper directement des lignes de commandes sql
- avec des queries

# Avec les décorateurs 

```python
import csv, sqlite3, pandas

con = sqlite3.connect("liste_eleve.sqlite")
cur = con.cursor()
```

```python
%load_ext sql
```

```python
#Si la base est dans un dossier data : %sql sqlite:///data/liste_eleve.sqlite
%sql sqlite:///liste_eleve.sqlite
```

```python
%%sql 
SELECT Nom FROM liste_eleve;
```

```python
%%sql
CREATE VIEW vue_eleve AS
SELECT Nom, Prénom, "Né(e) le"
FROM liste_eleve;
```

```python
%%sql
SELECT name FROM sqlite_master WHERE type='view';
```

```python
%%sql
SELECT * FROM vue_eleve;
```

# Avec les Queries

```python
# Exécuter une requête SQL et stocker le résultat dans un DataFrame pandas
import csv, sqlite3, pandas

query = "SELECT * FROM liste_eleve"
df = pd.read_sql(query, con)

# Afficher DataFrame
print(df)
```
