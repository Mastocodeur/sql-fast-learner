```python
%%sql
SELECT name, elevation FROM volcans ORDER BY elevation DESC LIMIT 10 ;
```

```python
%%sql
SELECT COUNT(name) FROM volcans WHERE elevation > 5000 
```

```python
%%sql
SELECT name, elevation, wiki FROM volcans WHERE elevation < 300 ORDER BY elevation DESC
```

```python
%%sql
SELECT name, eruption_year
FROM volcans
WHERE eruption_year > 2000;
```

```python
%%sql
SELECT COUNT(*)
FROM volcans
WHERE eruption_year IS NOT NULL;
```

```python
%%sql
SELECT name, eruption_year
FROM volcans
ORDER BY eruption_year DESC;
```

```python
%%sql SELECT name, lat, lon
FROM volcans
WHERE lat < 0;
```

```python
%%sql
SELECT COUNT(*)
FROM volcans
WHERE lat > 0 AND lat < 30;
```

```python
%%sql
SELECT name, lat
FROM volcans
ORDER BY lat DESC
LIMIT 5;
```

```python
%%sql 
SELECT name, abstract
FROM volcans
WHERE abstract LIKE '%éruption%';
```
