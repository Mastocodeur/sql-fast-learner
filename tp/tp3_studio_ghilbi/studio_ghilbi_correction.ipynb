4. Recommencer la même sélection, mais en faisant un tri croissant sur l'âge des personnages.

```python
%%sql
SELECT nom, age
FROM personnage
ORDER BY age ASC;
```

- Vérifier que le personnage le plus jeune avec un âge connu est «Mei Kusakabe».
- 
Quelle remarque peut-on faire ?- 
Comment expliquer ce résultat ? Comment aurait-on pu éviter le problème ?

```python
%%sql
SELECT nom, age
FROM personnage
ORDER BY age ASC
LIMIT 1;
```

Ce résultat est probablement dû à la façon dont les valeurs NULL ou vides sont traitées dans SQL. Par défaut, les valeurs NULL ou les chaînes vides peuvent être considérées comme les "plus petites" valeurs lors du tri. Ainsi, lors d'un tri ascendant, un enregistrement sans valeur d'âge ou avec une chaîne vide pourrait apparaître en premier, même si ce n'est pas l'intention.

Pour éviter ce problème et s'assurer que seuls les personnages avec un âge connu et valide sont pris en compte, il faut explicitement exclure les enregistrements sans âge défini (c'est-à-dire NULL ou vides) et ceux avec des âges textuels non numériques. Voici comment procéder :

Exclure les valeurs NULL et les chaînes vides : En utilisant une clause WHERE pour filtrer les âges valides.
Vérifier la validité des âges : S'assurer que les âges sont des valeurs numériques valides.

```python
%%sql
SELECT nom, age
FROM personnage
WHERE age IS NOT NULL AND age != '' 
AND age REGEXP "^[0-9]+$"
ORDER BY CAST(age AS UNSIGNED) ASC
LIMIT 1;
```
